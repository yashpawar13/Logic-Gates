# BASIC LOGIC GATES

### + OR GATE

```python
def OR(A,B):
    if A == 0 and B ==0:
        return 0
    else:
        return 1
```

Manual Print

```python
print ("Entre Bollean 1 or 0: ")
A = int(input ("A = "))
B = int(input ("B = "))
print('Q =',OR(A,B))
```

Automated Print

```python
from prettytable import PrettyTable
table = PrettyTable
table.field_names = ['A','B','Q']
for A in range(2):
    for B in range(2):
        Q = OR(A,B)
    table.add_row([A,B,Q])
print("OR Table")
print(table)
```
